```
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```
#import csv of goals scored and conceeded 
#https://www.whoscored.com/Regions/252/Tournaments/2/Seasons/8228/Stages/18685/TeamStatistics/England-Premier-League-2020-2021

scoreddf = pd.read_csv('goalsscored.csv')
concdf = pd.read_csv('goalsconceeded.csv')
```

```
#ranking teams in order of situational statistics
scoreddf['Openplay_rank'] =  scoreddf['Open Play'].rank(method='min', ascending = False)
scoreddf['Counter_rank'] =  scoreddf['Counter Attack'].rank(method='min', ascending = False)
scoreddf['Setpiece_rank'] =  scoreddf['Set Piece'].rank(method='min', ascending = False)
scoreddf['Penalty_rank'] =  scoreddf['Penalty'].rank(method='min', ascending = False)

concdf['Openplay_rank'] =  concdf['Open Play'].rank(method='min')
concdf['Counter_rank'] =  concdf['Counter Attack'].rank(method='min')
concdf['Setpiece_rank'] =  concdf['Set Piece'].rank(method='min')
concdf['Penalty_rank'] =  concdf['Penalty'].rank(method='min')
```

```
#set index of dataframe to the team name

scoreddf = scoreddf.set_index('Team')
concdf = concdf.set_index('Team')
```

```
#drop non-useful column 

scoreddf = scoreddf.drop(['R'], axis = 1)
concdf = concdf.drop(['R'], axis = 1)
```

```
#function to compare two selected teams
#if one is particularly stronger than another in a statistic (e.g. set piece goals 
#   conceeded), it will print out

def mismatch(team1, team2): 
    #check advantages for team 1
    stats1, stats2 = scoreddf.loc[team1][5:], concdf.loc[team2][5:]
    print(f'{team1} better at: ', end=' ')
    for x in range(len(stats1)):
        if (stats1[x] - stats2[x]) <= -10: #difference in rank is at least 10
            print(concdf.loc[team1][5:].index[x], end=' ')

    #check advantages for team 2
    stats1, stats2 = scoreddf.loc[team2][5:], concdf.loc[team1][5:]
    print(f'\n{team2} better at: ', end=' ')
    for x in range(len(stats1)): 
        if (stats1[x] - stats2[x]) <= -10: #difference in rank is at least 10
            print(concdf.loc[team1][5:].index[x], end=' ')
```

```
#relevant fixtures to compare

fixtures = ['Crystal Palace', 'Leicester', 
            'Chelsea', 'Aston Villa', 
            'Everton', 'Manchester City', 
            'Brighton', 'Arsenal', 
            'Burnley', 'Sheffield United', 
            'Southampton', 'West Ham', 
            'West Bromwich Albion', 'Leeds', 
            'Manchester United', 'Wolverhampton Wanderers', 
            'Tottenham', 'Fulham', 
            'Newcastle United', 'Liverpool', ]
```

```
#go through each fixture and compare the teams

counter = 0
while counter <= 19: 
    team1, team2 = fixtures[counter], fixtures[counter+1]
    #print(team1, team2)
    mismatch(team1, team2)
    print('\n')
    counter +=2 
```
